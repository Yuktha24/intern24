```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
data=pd.read_csv("/dataset - netflix1 (1).csv",na_values="Not Given")
print(data)
```

```python
grpyear = data.groupby(["release_year","type"],as_index = False).size()
grpyear = grpyear[grpyear.release_year >= 2010]

sns.lineplot(x="release_year", y="size", hue="type", data=grpyear)
```

```python
types = data['type'].value_counts()
labels = ['type', 'listed_in']
colors = ['#8464a0','#cea9bc']
plt.pie(types.values, labels = labels, autopct= "%3.3f%%", colors = colors);
plt.legend();
plt.title('Shows on netflix')
```

```python
text = data.listed_in[1]

wordcloud = WordCloud().generate(text)
word_string=" ".join(data['listed_in'].str.lower())

plt.figure(figsize=(10,10))
wc = WordCloud(background_color="violet", stopwords = STOPWORDS, max_words=1490, max_font_size= 300,  width=1600, height=800)
wc.generate(word_string)

plt.imshow(wc.recolor( colormap= 'jet' , random_state=17), interpolation="bilinear")
plt.axis('off')
plt.show()
```
